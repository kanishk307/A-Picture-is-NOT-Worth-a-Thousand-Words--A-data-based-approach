```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('googleCaptionDataset.csv',index_col=None)
```

```python
df.columns = ['Caption','ImgUrl']
```

```python
df['Caption']
```

```python
from collections import Counter
EveryWordCount = Counter()
df['Caption'].str.lower().str.split().apply(EveryWordCount.update)
```

```python
print(EveryWordCount)
```

```python
df['Caption'] = df['Caption'].str.replace('.','')
```

```python
df['TotalWords'] = df['Caption'].str.split().str.len()
```
